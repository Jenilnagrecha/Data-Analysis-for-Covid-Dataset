```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# read .csv file
df = pd.read_csv('Total_Cases_of_States.csv')
new_df1 = pd.read_csv('usa_final.csv')
citydata =pd.read_csv('Total_Cases_of_Cities.csv')
state1 = pd.read_csv('Total_Cases_of_States.csv')
final3 = pd.read_csv('Statewise_cases_on_each_day.csv')
```

```python
p = new_df1.iloc[:,0].values
q = new_df1.iloc[:,3].values
q1= new_df1.iloc[:,4].values


plt.rcParams["figure.figsize"] = (6.4,4.8)
plt.plot(p,q, color = 'blue', label = 'Daily Covid Cases')
plt.plot(p,q1, color = 'red', label = 'Daily Covid Deaths')
plt.title('Date vs No. of Cases', fontsize = 15)
plt.xlabel('Date', fontsize = 13)
plt.ylabel('Number of Cases',fontsize = 13)
plt.xticks(rotation= 90)
plt.legend()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (15,7)
p = final3.iloc[0:743,5].values
q = final3.iloc[0:743,1].values
plt.title("Alabama State")
plt.xlabel("Date")
plt.ylabel("Daily Covid Cases")
plt.plot(q,p)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (150,10)
plt.plot(p,q, color = 'blue')
plt.title('Date vs New Covid Cases', fontsize = 23)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('New Covid Cases',fontsize = 15)
plt.xticks(rotation= 70)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (150,10)
plt.plot(p,q1, color = 'red')
plt.title('Date vs New Covid Deats', fontsize = 23)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('New Covid Deaths', fontsize = 15)
plt.xticks(rotation= 70)
plt.show()
```

```python

citydata.head(100)
cityofalbama = citydata["Cities"].head(8).tolist()

fixedx = cityofalbama

death_y = citydata["Deaths"].head(8).tolist()
confirmed_y = citydata["Confirmed"].head(8).tolist()

x = lambda c : c/100
confirmed_y = list(map(x,confirmed_y))
plt.rcParams["figure.figsize"] = (6.4,4.8)
plt.bar(fixedx,death_y,align = 'edge',label = "death",width = .3,color = "#123456")
plt.bar(fixedx,confirmed_y,align = 'edge',width = (-.3),label = "confirmed")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (6.4,4.8)
deaths = citydata["Deaths"].head(68).tolist()
plt.hist(deaths,bins=50,edgecolor="#0e0afa",color="#fa560a")
plt.tight_layout()
plt.show()
```

```python
state = state1.iloc[:,1:4].values
x = state[:,0]
y = state[:,1]
z = state[:,2]

plt.rcParams["figure.figsize"] = (29,5)
plt.plot(x, y,label = 'Confirmed')
plt.plot(x, z,color ='red', label = 'Deaths')

plt.title('State vs Cases')
plt.xlabel('States')
plt.ylabel('Number')
plt.xticks(rotation= 70)
plt.legend(prop={'size':15})
plt.show()
```
