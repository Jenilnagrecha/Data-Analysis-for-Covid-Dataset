**Importing Libraries and Dataset**

---



```python
import pandas as pd
import numpy as np
import tabulate

df = pd.read_csv('us_simplified.csv')
print(df.head(100))
```

**Data Cleaning**

---



```python
df = df.drop('Country/Region', 1)
print(df.head(20).to_string())
```

```python
df.rename(columns = {'Admin2' : 'Cities'}, inplace = True)
df
```

```python
df = df[df.Cities != "Unassigned"]
df1 = df[df['Cities'].notna()].copy()
df0 = df1.copy()
df1
```

```python
df1 = df1.reset_index(drop=True)
df1
```

**Data Preprocessing**

---









New Columns for Confirmed Cases on particular day

```python
df1['Confirmed_on_day'] = abs(df1['Confirmed'].shift(1) - df1['Confirmed'])
df1['Deaths_on_day'] = abs(df1['Deaths'].shift(1) - df1['Deaths'])
i = 744
while i < len(df1):
    df1.loc[i, "Confirmed_on_day"] = 0
    df1.loc[i, "Deaths_on_day"] = 0
    i = i + 744
df1.loc[0, "Confirmed_on_day"] = 0
df1.loc[0, "Deaths_on_day"] = 0
df1
```

Extracting Total Cases in all Cities

```python
df2 = pd.DataFrame(columns=['Cities','Province/State','Confirmed','Deaths'])
i = 743
count = 0
while i < len(df1):
    df2.loc[count, "Province/State"] = df1.loc[i, "Province/State"]
    df2.loc[count, "Confirmed"] = int(df1.loc[i, "Confirmed"])
    df2.loc[count, "Deaths"] = int(df1.loc[i, "Deaths"]) 
    df2.loc[count, "Cities"] = df1.loc[i, "Cities"]
    i = i + 744
    count = count + 1
    
df2.to_csv('Total_Cases_of_Cities.csv')
df2
```

Extracting Total Cases in all States

```python
df3 = df2.drop('Cities', 1)
df3.to_csv('temp.csv')
df4 = pd.read_csv('temp.csv')
df4 = df4.groupby(['Province/State']).sum()
df4 = df4.drop('Unnamed: 0',1)
df4.to_csv('Total_Cases_of_States.csv')
df4.head(15)
```
